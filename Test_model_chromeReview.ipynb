```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import string
import spacy

from matplotlib.pyplot import imread
from wordcloud import WordCloud,STOPWORDS
%matplotlib inline
import re
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
p = PorterStemmer()
all_stopwords= stopwords.words('english')
all_stopwords.remove('not')
import warnings
warnings.filterwarnings("ignore")
```

```python
dataset= pd.read_csv('Test_Chrome_Reviws.csv')
dataset
```

```python
dataset=dataset.drop(['ID','Review URL','Thumbs Up','User Name','Developer Reply','Developer Reply','Version','Review Date','App ID'],axis=1)
dataset
```

```python
corpus=[]
n=len(dataset)
for i in range(0,n):
    sent = re.sub('[^a-zA-Z]', ' ',str(dataset['Text'][i]))
    sent = sent.lower()
    sent = sent.split()
    sent = [p.stem(word) for word in sent if not word in set(all_stopwords)]
    sent = ' '.join(sent)
    corpus.append(sent)
```

```python
import pickle
cv = pickle.load(open('chrome_review_pridiction.pkl', "rb"))
```

```python
X_fresh = cv.transform(corpus).toarray()
X_fresh.shape
```

```python
import joblib
classifier = joblib.load('Classifier_Sentiment_Model')
```

```python
y_pred = classifier.predict(X_fresh)
print(y_pred)
```

```python
y_pred.shape
```

```python
dataset['predicted_label'] = y_pred.tolist()
dataset.head(20)
```

```python
dataset.to_csv("Predicted_Reviews_Fresh_Dump.tsv", sep='\t', encoding='UTF-8', index=False)
```
